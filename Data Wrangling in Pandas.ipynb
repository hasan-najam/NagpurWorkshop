```python
import numpy as np
def get_x_coordinate(series):
    a=series.RA
    d=series.Dec
    return np.cos(a)*np.cos(d)

df.apply(get_x_coordinate,axis=1)
#x = cos(a) cos(d)
#a is for RA and d is DEC
#y = sin(a) cos(d)
#z = sin(d) 
```

```python
def get_x_coordinate(series):
    print('The Series is:')
    print(type(series))
    print(series)
    a=series.RA
    print('The Extracted RA is:')
    print(a)
    d=series.Dec
    print('The Extracted Dec is')
    print(d)
    print('----------------')
    return np.cos(a)*np.cos(d)

df.head().apply(get_x_coordinate,axis=1)
```
